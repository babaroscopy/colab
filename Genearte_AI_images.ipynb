```python
# prompt: write code or model to generate images on a topic and save it in folder? let's say 100 images on each topic. images should be ads so topics should be accordingly

from diffusers import StableDiffusionPipeline
import torch
import os
from PIL import Image

# Initialize Stable Diffusion pipeline
model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16) # Use float16 for potential memory savings
pipe.to(device)


# Define output folder
output_folder = "/content/generated_images"
os.makedirs(output_folder, exist_ok=True)

# Prompts for image generation (more ad-specific examples)
prompts = [
  #  "A futuristic sports car, sleek design, vibrant neon lights, advertising tagline: 'Experience the Future of Driving'",
   # "A group of diverse people enjoying a picnic in a lush park, high-quality product shot of gourmet food, advertising tagline: 'Taste the Moments'",
    #"Close-up shot of a smartphone with a stunning display, highlighting its camera features, advertising tagline: 'Capture Every Detail'",
    "A family laughing together on a beach vacation, with a clear shot of comfortable and stylish beachwear, advertising tagline: 'Make Memories, Not Excuses'",
    "A person using a laptop outdoors in a coffee shop, showcasing its powerful performance, advertising tagline: 'Work Seamlessly, Anywhere'",
    "A person exercising with a stylish smartwatch, showing its health tracking features, advertising tagline: 'Track Your Fitness Journey'",
    "High-quality shot of a luxurious watch, highlighting its craftsmanship, advertising tagline: 'Timeless Elegance'",
    "A group of friends enjoying a refreshing beverage together, in a vibrant social setting, advertising tagline: 'Refresh Your Spirit'",
    "A close-up shot of sustainable and eco-friendly cleaning products, showcasing their natural ingredients, advertising tagline: 'Clean Your Conscience'",
    "A person working from home with a comfortable and ergonomic office chair, advertising tagline: 'Work Comfortably, Work Effectively'"
]

# Generate multiple images per prompt
images_per_prompt = 10 # Reduced for demonstration; change back to 100

for prompt_idx, prompt in enumerate(prompts):
    prompt_folder = os.path.join(output_folder, f"prompt_{prompt_idx+1}")
    os.makedirs(prompt_folder, exist_ok=True)

    for i in range(images_per_prompt):
        try:
          # Generate image with a seed for reproducibility (optional)
          generator = torch.Generator(device=device).manual_seed(i)
          image = pipe(prompt, generator=generator).images[0]
          image_path = os.path.join(prompt_folder, f"image_{i+1}.png")
          image.save(image_path)
          print(f"Generated image saved: {image_path}")
        except Exception as e:
          print(f"Error generating image for prompt {prompt_idx+1}, image {i+1}: {e}")
```
